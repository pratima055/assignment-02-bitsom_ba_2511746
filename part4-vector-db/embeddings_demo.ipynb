This notebook demonstrates how sentence embeddings capture semantic meaning using the all-MiniLM-L6-v2 model. It includes similarity analysis and query matching.

## Step 1: Define Sample Sentences

We create 10 sentences across three topics: Cricket, Cooking, and Cybersecurity.

```python
sentences = [
    "[Cricket] The batsman scored a century in the final match.",
    "[Cricket] The bowler took a wicket with a perfect yorker.",
    "[Cricket] The team celebrated their victory after winning the series.",
    "[Cricket] The captain set an aggressive field placement.",

    "[Cooking] The chef prepared a delicious pasta with fresh ingredients.",
    "[Cooking] Baking a cake requires precise measurements and timing.",
    "[Cooking] She added spices to enhance the flavor of the curry.",

    "[Cybersecurity] Strong passwords are essential for protecting online accounts.",
    "[Cybersecurity] Encryption helps secure sensitive data from hackers.",
    "[Cybersecurity] Firewalls prevent unauthorized access to networks."
]
```

## Step 2: Generate Sentence Embeddings

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(sentences)
```

## Step 3: Compute Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity_matrix = cosine_similarity(embeddings)
```

## Step 4: Visualize Similarity Matrix

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))
sns.heatmap(similarity_matrix, annot=True, cmap='coolwarm')
plt.title('Sentence Similarity Matrix')
plt.show()
```

## Step 5: Query Matching

```python
query = 'The bowler took three wickets in one over'

query_embedding = model.encode([query])
similarities = cosine_similarity(query_embedding, embeddings)[0]
```

## Step 6: Retrieve Most Similar Sentences

```python
import numpy as np

top_indices = np.argsort(similarities)[-2:][::-1]

print('Query:', query)
print('\nTop 2 similar sentences:\n')

for idx in top_indices:
    print(f'Sentence: {sentences[idx]}')
    print(f'Similarity Score: {similarities[idx]:.4f}\n')
```

## Conclusion

The results show that sentences from the same topic have higher similarity scores. The query about a bowler correctly matches cricket-related sentences, demonstrating that embeddings capture semantic meaning beyond keyword matching.
